### Making changes to the data

```python
df.head(3)
```

```python
# Adding a column and adding values to it.
# This is easy way but inefficient and also slow.
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(3)
```

```python
# one more way to do the same.
# driopping the old column. or we can do inplace=True at the end of the drop function.
df = df.drop(columns=['Total'])
df.head(3)
```

```python
# Now adding a column in different way
df['Total'] = df.iloc[:,4:10].sum(axis=1)
# iloc decode [all the rows, from_columns: to_columns].sum(axis=1)
# .sum is an internal function to do add the data.
# if axis=0, it will add data horizontally. axis=0 will do vertically.
df.head(3)
```

```python
# Moving column to other different places in the sheet.
# df = df[['Total','HP','Defense']]  # one way way to do this. But iefficient again.
cols = df.columns.values
cols
```
