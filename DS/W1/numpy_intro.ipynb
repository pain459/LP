### Array Operations

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a - b
d = a * b

print(c)
print(d)
```

```python
f = np.array([0, -10, -5, -15, 0])
c = (f - 31) * (5/9)
c
```

```python
# boolean arrays
c > -20
```

```python
c % 2 == 0
```

```python
# matrix multiplication
A = np.array([[1, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
print(A * B)  # normal multiplication
print(A @ B)  # matrix multiplication
```

```python
A.shape
```

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7.1, 8.2, 9.3], [10.4, 11.3, 12.5]])
print(array1.dtype)
print(array2.dtype)
```

```python
array3 = array1 + array2
print(array3.dtype)
```

```python
print(array3.max())
print(array3.min())
print(array3.sum())
print(array3.mean())
```

```python
from PIL import Image
from IPython.display import display
im = Image.open('sample.tif')
display(im)
```

```python
array = np.array(im)
print(array.shape)
array
```

```python
mask = np.full(array.shape, 255)
mask
```

```python
modified_array = array - mask
```

```python
modified_array = modified_array * -1
```

```python
modified_array = modified_array.astype(np.uint8)
modified_array
```

```python
display(Image.fromarray(modified_array))
```

```python
reshaped = np.reshape(modified_array, (480, 640))
print(reshaped.shape)
display(Image.fromarray(reshaped))
```
