```python
import pandas as pd
df = pd.read_csv('datasets/presidents.csv')
df.head(10)
```

```python
df['First'] = df['President']
df['First'] = df['President'].replace("[ ].*", "", regex=True)
df.head()
```

```python
del(df['First'])
```

```python
def splitname(row):
    row['First']=row['President'].split(" ")[0]
    row['Last']=row['President'].split(" ")[-1]
    return row

df=df.apply(splitname, axis='columns')
```
