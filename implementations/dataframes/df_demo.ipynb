```python
n = 3 # number of records to drop
df2 = df.iloc[n:,:]
df2.reset_index(drop=True, inplace=True)
df2.head()
```

```python
# Setting up the new header for df2
new_header = df2.iloc[0] # grab a new header.
df2 = df2[1:] # take the data less the header.
df2.columns = new_header # set the new header as data header.
df2.reset_index(drop=True, inplace=True)
df2.head()
```

```python
df2.dtypes
```

```python
surface_mines = df2[df2['Mine Type'] == 'Surface']
surface_mines.head()
```

```python
surface_active = df2[(df2['Mine Status'] == 'Active') & (df2['Mine Type'] == 'Surface')]
surface_active.head()
```

```python
surface_active.shape[0]  # .shape offers (columns, rows) information. You can filter them.
```

```python
surface_active_employees = df2[(df2['Mine Status'] == 'Active') & (df2['Mine Type'] == 'Surface') & (df2['Average Employees'] >= 50)]
surface_active_employees.shape[0]
```

```python
filtered_subsidary = surface_active_employees[surface_active_employees['Company Type'].str.contains('Subsidiary', case=False, na=False)]
filtered_subsidary.head()
```
